```python
df = netflix.copy()

df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)

df.dropna(subset=['date_added', 'rating', 'duration'], inplace=True)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), format="%B %d, %Y", errors='coerce')
df['date_added']
```

```python
df['day_added'] = df['date_added'].dt.day#Convert String to time for analysis
df['month_added'] = df['date_added'].dt.month
df['year_added'] = df['date_added'].dt.year
df['year_added']=df['year_added'].astype(int)
df['day_added'] = df['day_added'].astype(int)
df['month_added'] = df['month_added'].astype(int)
```

```python
plt.figure(figsize=(12, 8))  
sns.countplot(y='rating', hue='type', data=netflix, order=netflix['rating'].value_counts().index)
plt.title('Content Rating Distribution by Type')
plt.show()#Showing Distribution of Rating Per Type
```

```python
#Show Rating of Movies and Shows
sns.countplot(df['rating'],color='red')
fig = plt.gcf()
fig.set_size_inches(12,12)
plt.title("Rating")
```

```python
plt.figure(figsize=(12, 8))  
age_groups = {
    'Kids': ['TV-Y', 'TV-Y7', 'G', 'PG'],
    'Teens': ['TV-PG', 'PG-13'],
    'Adults': ['TV-14', 'R', 'TV-MA', 'NC-17']
}

def age_group(rating):
    for group, values in age_groups.items():
        if rating in values:
            return group
    return 'Other'

netflix['age_group'] = netflix['rating'].apply(age_group)
sns.countplot(x='age_group', hue='type', data=netflix)
plt.title('Target Age Group by Type')
plt.show()
##Showing Target Age per type
```

```python
labels = ['Movie', 'TV show']
size = netflix['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
#Showing Distribution of Type (TV Show , Movie)
```

```python
top_countries = netflix['country'].value_counts().head(10)
sns.barplot(y=top_countries.index, x=top_countries.values)
plt.title('Top 10 Countries by Number of Titles')
plt.xlabel('Number of Titles')
plt.show()
```

```python
from collections import Counter

all_genres = netflix['listed_in'].dropna().apply(lambda x: x.split(', '))
genre_counts = Counter([genre for sublist in all_genres for genre in sublist])
top_genres = dict(genre_counts.most_common(10))

sns.barplot(x=list(top_genres.values()), y=list(top_genres.keys()))
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Number of Titles')
plt.show()
```

```python
movies = netflix[netflix['type'] == 'Movie']
movies['duration_minutes'] = movies['duration'].str.extract('(\d+)').astype(float)

plt.hist(movies['duration_minutes'], bins=30, color='skyblue')
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Count')
plt.show()
```

```python
from collections import Counter

all_directors = netflix['director'].dropna().apply(lambda x: x.split(', '))
flat_directors = [d for sublist in all_directors for d in sublist]
top_directors = Counter(flat_directors).most_common(10)

df_top_directors = pd.DataFrame(top_directors, columns=['Director', 'Count'])
sns.barplot(y='Director', x='Count', data=df_top_directors)
plt.title('Top 10 Directors on Netflix')
plt.show()
#Showing Top 10 Directors on netflix
```

```python
plt.figure(figsize=(12, 6))  

netflix['year_added'].value_counts().sort_index().plot(kind='bar', color='skyblue')

plt.title('Number of Titles Added Per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=0)  
plt.tight_layout()       
plt.show()
```
